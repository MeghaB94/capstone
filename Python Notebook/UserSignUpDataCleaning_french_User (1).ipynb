```python
from googletrans import Translator

# Load your DataFrame
# translated_df = pd.read_csv('your_translated_file.csv') # Assuming you have already loaded it

# Initialize the translator
translator = Translator()

# Define a function to translate text
def translate_text(text):
    if pd.isna(text):  # Check for NaN values
        return text
    try:
        translated = translator.translate(text, src='fr', dest='en')
        return translated.text
    except Exception as e:
        print(f"Error translating text: {text}\n{e}")
        return text

# Apply the translation function to the entire DataFrame
translated_df = translated_df.applymap(translate_text)

# Display the translated DataFrame to verify
translated_df.head()

# # Optionally, save the translated DataFrame to a CSV file
# translated_df.to_csv('translated_to_english.csv', index=False)
```
