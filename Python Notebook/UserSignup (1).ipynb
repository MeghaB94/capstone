```python
import pandas as pd
from googletrans import Translator

# Original DataFrame with French column names
original_df = french_df.copy()  # Replace "your_data.csv" with the path to your CSV file

# French column names
french_columns = ['Username', 'FirstName', 'LastName', 'EmailAddress', 'Phone', 'City',
                  'Province', 'Country', 'Highest level of education', 'Cohort Name',
                  'Program Version', 'Genre', 'Âge', 'Statut au Canada',
                  'Situation maritale', 'Pays de naissance',
                  'Pays doù vous avez immigré immigrerez',
                  'Année darrivée au Canada saisir 0 si non applicable',
                  'Principale langue parlée à la maison', 'Plus haut niveau de formation',
                  'Quelle est votre situation professionnelle actuelle ',
                  'Dans quelle industrie ou secteur travaillez vous',
                  'Est ce votre industrie ou secteur de prédilection',
                  'Quel est votre salaire annuel actuel', 'Organization', 'DateAdded',
                  'Quarterly']

# Translate French column names to English
translator = Translator()
english_columns = [translator.translate(col, src='fr', dest='en').text for col in french_columns]

# Rename columns in original DataFrame
original_df.columns = french_columns

# Create DataFrame with translated column names and copy data
translated_df = original_df.copy()
translated_df.columns = english_columns
```
