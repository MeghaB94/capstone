Combining all Dataframes

```python
df_signup = pd.concat([df_signupJan24, df_signupJuly23, df_signupOct23], ignore_index=True)
```

```python
df_signup.to_csv('data_signup.csv', index=False)
```

```python
print(df_signup.isnull().sum())
```

```python
# Filter the DataFrame where columns 65, 66, and 69 have missing values
missing_subset = df_signup[df_signup[['Gender', 'What is your current employment status', 'Country you have_will immigrate from']].isnull().any(axis=1)]

print(missing_subset)
```

```python
print("Row numbers of the missing subset:")
print(missing_subset.index)
```

```python
rows_to_drop = missing_subset.index

df_cleaned = df_signup.drop(index=rows_to_drop)

print("Shape of DataFrame after dropping rows:", df_cleaned.shape)
```

**VISUALIZATIONS**

```python
plt.figure(figsize=(10, 6))
df_signup['Quarterly'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Quarterly Comparison of Signups')
plt.xlabel('Quarter')
plt.ylabel('Number of Signups')
plt.xticks(rotation=45)
plt.show()
```

```python
df_signup_sorted = df_signup.sort_values(by='Age')
plt.figure(figsize=(8, 6))
sns.histplot(df_signup_sorted['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
gender_counts = df_signup['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_signup, x='What is your current employment status')
plt.title('Count of Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(y=df_signup['Country'], order=df_signup['Country'].value_counts().index)
plt.title('Distribution of Signups by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```
